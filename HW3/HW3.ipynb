### Q6

```python
def F_inv(u):
    c = 1 - math.exp(-0.05)
    return math.log(1/(1-c*u))

u = np.random.rand()
x = F_inv(u)
print(u, x)
```

```python
ans = 0
num_trials = 1000
for _ in range(num_trials):
    u = np.random.rand()
    x = F_inv(u)
    ans += x
ans = ans/num_trials
print(ans)
```

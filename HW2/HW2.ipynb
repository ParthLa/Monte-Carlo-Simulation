```python
def simulate_integral_value(k, h):
    ans = 0
    for _ in range(k):
        y = np.random.rand() # U[0, 1]
        ans += h(y)
    ans = ans/k
    return ans
```

```python
# Program to generate n values from the pmf with p1 = 1/3 and p2 = 2/3
p1 = 1/3
def simulate_pmf12(n):
    output = []
    v = np.random.rand(n)
    for val in v:
        if(val<=p1):
            output.append(1)
        else:
            output.append(2)
    output = np.array(output)
    return output
```

```python
for n in [100, 1000, 10000]:
    output = simulate_pmf12(n)
    assert(len(output) == n)
    prop = len(output[output == 1])/len(output)
    print(n, prop)
```

### Q2

```python
def simulate_pmf_n(pvec) -> float:
    '''
    Args:
        pvec: vector of probabilities defining the probability mass function (pmf)
    Returns:
        value of the Random Variable following the given pmf
    '''
    n = len(pvec)
    v = np.random.rand()
    # The cdf value (upto index i; i ranging from 1 to n)
    F = p1
    ans = -1
    for i in range(1, n+1):
        if(v<=F):
            ans = i
            break
        if(i==n):
            raise Exception("index of iteration should not reach n!")
        F += pvec[i]
    assert(ans != -1)
    return ans
```

### Q4

```python
n = 100
v = []

num_trials = 50000
for _ in range(num_trials):
    # seq will represent the sequence of shuffled cards

    # Creating a base sequence
    seq = np.array([i for i in range(n+1)])
    
    # Shuffling the sequence
    for k in range(n, 1, -1):
        u = np.random.rand()
        index = int(k*u) + 1
        
        # swapping seq[index] and seq[k]
        temp = seq[k]
        seq[k] = seq[index]
        seq[index] = temp

    # counting the number of hits
    cnt = 0
    for i in range(1, len(seq)):
        if(seq[i] == i):
            cnt += 1
    v.append(cnt)

expectation = np.mean(v)
variance = np.var(v) 
print(expectation, variance)
```

- Expectation: 1
- Variance: 1

```python
num_trials = int(2e5)
ans = 0
for _ in range(num_trials):
    # cnt represents the number of times we throw the pair of dice; 
    #   number of throws required to get all possible sums atleast once
    cnt = 0
    # s stores only the distinct values; so once s has all the elements from 2 to 12 
    #   (11 possible values), we are guaranteed to have each sum atleast once
    #   Note that no other value (of the sum) is possible; 
    #   so len(s) = 11 <=> all possible sums (from 2 to 12) occured atleast once 
    s = set([])
    while(len(s) != 11):
    # while(len(s) != 6):
        # simulate throwing of 2 dice
        cnt += 1
        # get the value
        u1 = np.random.rand()
        v1 = int(6*u1) + 1
        u2 = np.random.rand()
        v2 = int(6*u2) + 1
        # v2 = 0
        s.add(v1+v2)
    ans += cnt
ans = ans/num_trials
print(ans)
```
